## Series、DataFame、Panel

````python
import pandas as pd
````

````python
dataframe_list = pd.read_html("https://pandas.pydata.org/pandas-docs/stable/reference/frame.html")
dataframe_df = pd.concat(dataframe_list, ignore_index=True)

series_list = pd.read_html("https://pandas.pydata.org/pandas-docs/stable/reference/series.html")
series_list.pop(13)
series_df = pd.concat(series_list, ignore_index=True)
````

````python
df_df = dataframe_df.copy(deep=True)
df_df['df'] = dataframe_df[0].apply(lambda x: x.split('.')[-1])
df_df['func'] = df_df['df'].apply(lambda x: x.split('(')[0])
df_df = df_df[['func', 'df', 1]]
df_df.set_index('func', inplace=True)
df_df.to_csv('df.csv', sep='|', index=False)
````

````python
ss_df = series_df.copy(deep=True)
ss_df['ss'] = ss_df[0].apply(lambda x: x.split('.')[-1])
ss_df['func'] = ss_df['ss'].apply(lambda x: x.split('(')[0])
ss_df = ss_df[['func', 'ss', 1]]
ss_df.set_index('func', inplace=True)
ss_df.to_csv('ss.csv', sep='|', index=False)
````

````python
ss_df.join(df_df, how='outer', lsuffix='_ss', rsuffix='_df', sort=True)[['ss', 'df']]
````
