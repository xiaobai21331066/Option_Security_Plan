# 回测结果解读

回测收益率：策略在期限内的收益率
$$回测收益率=\frac{期末投资组合总权益 - 期初投资组合总权益}{期初投资组合总权益}$$

年化收益率：采用了复利累积的年化方式计算得到的年化收益
$$年化收益率=(1 + R)^{\frac{1}{t}} - 1$$
其中：$R=累计收益率$, $t = \frac{策略运行的交易日}{252}$

阿尔法：CAPM模型中表达式的残差项，
